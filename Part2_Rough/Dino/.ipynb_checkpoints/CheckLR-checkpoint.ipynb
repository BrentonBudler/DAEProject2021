```python
test1 = df[["age","gender","race","province_current", "tertiary_edu","lost_labour","usual_work"]]
```

```python
test1.isnull().sum()
```

```python
test1 = test1.dropna()
```

```python
test1['gender'] = test1['gender'].map({'Man':1, 'Woman':0})
test1['gender'] = test1['gender'].astype(int)
```

```python
test1['tertiary_edu'] = test1['tertiary_edu'].map({'Yes':1, 'No':0})
test1['tertiary_edu'] = test1['tertiary_edu'].astype(int)
```

```python
test1
```

```python
test1 = pd.get_dummies(test1, columns=["race","province_current", "usual_work"], drop_first=True)
```

```python
sns.countplot(data = df, x = 'lost_labour')
plt.show() #Balanced as fuck 
```

```python
test1
```

```python
X = test1.drop('lost_labour',axis=1)
y = test1['lost_labour']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(max_iter=10000)
log_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = log_model.predict(X_test)
accuracy_score(y_test, y_pred)  
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(log_model,X_test, y_test)
plt.show()
```
