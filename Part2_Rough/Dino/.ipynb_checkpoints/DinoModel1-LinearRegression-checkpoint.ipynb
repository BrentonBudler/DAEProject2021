```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

#Linear Regression stuff
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dftemp = pd.concat([dem,lab],axis=1)
```

```python
df_main = dftemp[["age","gender","race","highest_grade","province_current", "tertiary_edu","take_home_pay_feb", "take_home_pay_apr","kept_labour","usual_work"]]
```

# Drop all nans

```python
display(df_main)
df_main_dropped_nans = df_main.dropna()
display(df_main_dropped_nans)
```

# Check salaries and difference

### Calculate and store salary difference

```python

diff = df_main_dropped_nans['take_home_pay_apr'] - df_main_dropped_nans['take_home_pay_feb']

display(diff)

perc = (diff/df_main_dropped_nans['take_home_pay_feb'])* 100

perc = np.nan_to_num(perc)
display(perc)

df_main_diff = df_main_dropped_nans.copy()

df_main_diff['take_home_pay_difference'] = perc
```

# Encode some vriables and get dummies

```python
df_main_removedinf['gender'] = df_main_removedinf['gender'].map({'Man':1, 'Woman':0})
df_main_removedinf['gender'] = df_main_removedinf['gender'].astype(int)
```

```python
df_main_removedinf['tertiary_edu'] = df_main_removedinf['tertiary_edu'].map({'Yes':1, 'No':0})
df_main_removedinf['tertiary_edu'] = df_main_removedinf['tertiary_edu'].astype(int)
```

```python
# test1[test1['kept_labour']==1]
```

```python
df_dummies = pd.get_dummies(df_main_removedinf, columns=["gender","race","highest_grade","province_current", "usual_work"], drop_first=True)
```

```python
display(df_dummies)
df_dummies.columns
```

```python
df_final = df_dummies.copy()
```

# Linear regression
Trying to predict change in salary

```python
y = df_final['take_home_pay_difference']

X = df_final.drop('take_home_pay_difference',axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


model = LinearRegression()
model.fit(X_train,y_train)

predicted = model.predict(X_test)


model.score(X_test.values, y_test.values)
```
