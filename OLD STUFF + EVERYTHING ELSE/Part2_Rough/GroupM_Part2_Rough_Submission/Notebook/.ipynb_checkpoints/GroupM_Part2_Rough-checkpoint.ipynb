```python
# df_main_kept_labour = df_main_diff[(df_main_diff['kept_labour']==1) & (df_main_diff['salary_change']!="None")]


df_main_kept_labour = df_main_dropped_nans[(df_main_dropped_nans['kept_labour']==1)]

print(list(df_main_kept_labour))

display(df_main_kept_labour)
```

### Calculate and store salary difference

```python

diff = df_main_kept_labour['take_home_pay_apr'] - df_main_kept_labour['take_home_pay_feb']

display(diff)

perc = (diff/df_main_kept_labour['take_home_pay_feb'])* 100

perc = np.nan_to_num(perc)
# display(perc)

df_main_diff = df_main_kept_labour.copy()

df_main_diff['take_home_pay_difference'] = perc
```

### Removing outliers

```python
print("Min percentage change before: ",min(df_main_diff['take_home_pay_difference']))
print("Max percentage change before: ", max(df_main_diff['take_home_pay_difference']))

df_main_removedinf = df_main_diff.drop(df_main_diff.index[df_main_diff['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])
df_main_removedinf = df_main_removedinf.drop(df_main_removedinf.index[df_main_removedinf['take_home_pay_difference'].argmax()])

print("\n\n\nMax percentage change after: ",max(df_main_removedinf['take_home_pay_difference']))
print("Min percentage change after: ",min(df_main_removedinf['take_home_pay_difference']))
```

### Determine if salary changed and create new variable for this

```python
df_newvars = df_main_removedinf.copy()


def salary_change(diff):
    if diff==0:
        return "unchanged"
    else:
        return "changed"
    

def salary_diff(diff):
    if diff==0:
        return "unchanged"
    elif diff>0:
        return "increased"
    elif diff<0:
        return "decreased"
    
    
def salary_decreased(diff):
    if diff<0:
        return "decreased"
    else:
        return "didn't decrease"
    
    
def salary_inc_dec(diff):
    if diff<0:
        return "decreased"
    elif diff>0:
        return "increased"
    else:
        return None
    

df_newvars['salary_change'] = np.vectorize(salary_diff)(df_newvars['take_home_pay_difference'])
```

### Encode some variables and get dummies

```python
df_grades_mapped = df_newvars.copy()

grade_map = {
    "Grade 12": 12,
    "Grade 11": 11, 
    "Grade 10": 10, 
    "Grade 9" : 9
}
df_grades_mapped.loc[:,'highest_grade'] = df_grades_mapped.highest_grade.map(grade_map)
df_grades_mapped = df_grades_mapped[df_grades_mapped.highest_grade.notnull()]
df_grades_mapped.highest_grade = df_grades_mapped.highest_grade.astype(int)
```

```python
df_encoded = df_grades_mapped.copy()
df_encoded['gender'] = df_encoded['gender'].map({'Man':1, 'Woman':0}).astype(int)
df_encoded['tertiary_edu'] = df_encoded['tertiary_edu'].map({'Yes':1, 'No':0}).astype(int)
```

```python
# df_dummies = pd.get_dummies(df_encoded, columns=["race","highest_grade","province_current", "usual_work"], drop_first=True)
df_dummies = pd.get_dummies(df_encoded, columns=["race", "usual_work"], drop_first=True)


display(df_dummies)
print(df_dummies.columns)
```

```python
df_dummies[df_dummies["salary_change"]=="increased"]
```

### Store everything done in a new dataframe and drop columns that we don't want in model

```python
df_final = df_dummies.copy()

# df_final = df_final.drop(['take_home_pay_difference'], axis=1)

# df_final = df_final.drop(['take_home_pay_difference', 'take_home_pay_feb', 'take_home_pay_apr'], axis=1)

df_final = df_final.drop(['take_home_pay_difference', 'take_home_pay_apr'], axis=1)

display(df_final)
```

### Check balance of classes that are being predicted

```python
sns.countplot(data = df_final, x = 'salary_change')
plt.show()
```
