```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_stata('Datasets/NIDS-CRAM_Wave3_Anon_V1.0.0.dta')
df.set_index('pid', inplace=True)
df
```

```python
""" 
Drop all unnecessary columns (ie: Columns that pertain to interviews from previous waves and flag responses):

w3_nc_outcome: Interview outcome
w3_nc_dob_m: Date of birth(month)
w3_nc_sample: From which sample did this respondent originate?
w3_nc_intrv_c: aa3 - Interviewer code
w3_nc_intrv_d: aa4_d - Date of interview (Day)
w3_nc_intrv_m: aa4_m - Date of interview (Month)
w3_nc_intrv_y: aa4_y - Date of interview (Year)
w3_nc_refexpl: aa5 - Reason for refusal
w3_nc_duration: Total duration of interview in minutes

w3_nc_popgrp_flg: ba2 flag - response surveyed or from historical data
w3_nc_mv2017_flg: ba9 flag - response surveyed or from historical data
w3_nc_mv2017: ba9 - Are you living in the same dwelling as when NIDS interviewed you in 2017?

w3_nc_intres: h1 - In gereral how did the respondent act towards you during the interview?
w3_nc_intlng1: h2_1 - Languages used during interview 1
w3_nc_intlng2: h2_2 - Languages used during interview 2
w3_nc_intlng3: h2_3 - Languages used during interview 3

"""

columns = ['w3_nc_outcome', 'w3_nc_dob_m', 'w3_nc_sample', 'w3_nc_intrv_c', 'w3_nc_intrv_d', 'w3_nc_intrv_m', 'w3_nc_intrv_y', 
           'w3_nc_refexpl', 'w3_nc_duration', 'w3_nc_popgrp_flg', 'w3_nc_mv2017_flg', 'w3_nc_mv2017','w3_nc_intres', 
           'w3_nc_intlng1', 'w3_nc_intlng2', 'w3_nc_intlng3']
df = df.drop(columns, axis=1)
df
```

```python
# Check to see which columns have the most NaN values, and decide if they should be removed due to a lack of info.

def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')
    
columns_missing = df.isnull().sum()
#display(columns_missing)
print_full(columns_missing)
```

```python
# Drop columns with more than 55% of values missing. 
# Should be left only with the columns that could provide some useful insights. 
df = df.dropna(thresh=0.55*len(df), axis=1)
display(df)
print(df.columns) #Get the names of all the remaining columns.
```
